#   使用股票代码列表下载多只股票数据，并将每只股票的数据保存到单独的CSV文件中

```python

# 定义股票代码列表
stock_symbols = ["300100", "300378", "300738", "600276", "688185"]
# stock_symbols = ["600276"]

# 定义日期范围
start_date = "20240101"
end_date = "20250314"

def clean_filename(name: str) -> str:
    """清理文件名中的非法字符"""
    # 替换Windows文件名禁用的字符 \ / : * ? " < > |
    return re.sub(r'[\\/*?:"<>|]', '_', name.strip())

# 遍历股票代码列表，下载数据并保存到CSV文件
for symbol in stock_symbols:
    try:
        # 获取股票信息
        info_df = ak.stock_individual_info_em(symbol=symbol)
               
        # 打印查看可用字段：print(info_df.columns.tolist())
        stock_name = info_df[info_df['item'] == '股票简称']['value'].values[0]
        
        # === 清理名称 ===
        cleaned_name = clean_filename(stock_name)
        
        # 下载股票数据
        stock_data = ak.stock_zh_a_hist(symbol=symbol, period="daily", start_date=start_date, end_date=end_date, adjust="qfq")
        
        # 保存数据到CSV文件
        csv_file = f"{symbol}_{stock_name}_historical_data.csv"
        stock_data.to_csv(csv_file, index=False,encoding="utf-8")
        print(f"Data for {symbol} saved to {csv_file}")
        
    except Exception as e:
        print(f"Failed to download data for {symbol}: {e}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 读取CSV数据
filename = '300100_双林股份_historical_data'
df = pd.read_csv(filename + '.csv', encoding='utf-8')

# 检查列名
print("列名:", df.columns)

# 检查数据的前几行
print("数据预览:\n", df.head())

# 确保列名正确
if '最低' not in df.columns:
    raise ValueError("列名 '最低' 不存在于数据中。请检查CSV文件的列名。")

# 转换日期格式，无法解析的日期转换为 NaT
df['日期'] = pd.to_datetime(df['日期'], errors='coerce')

# 检查是否有无法解析的日期
print("无法解析的日期:\n", df[df['日期'].isna()])

# 删除无法解析的日期行
df = df.dropna(subset=['日期'])

# 设置索引
df.set_index('日期', inplace=True)

# 计算MACD
def calculate_macd(data, fast_period=23, slow_period=50):
    ema_fast = data['收盘'].ewm(span=fast_period, adjust=False).mean()
    ema_slow = data['收盘'].ewm(span=slow_period, adjust=False).mean()
    macd = ema_fast - ema_slow
    return macd

# 计算随机值（%K）
def calculate_stochastic(data, period=3):
    low_min = data['最低'].rolling(window=period).min()
    high_max = data['最高'].rolling(window=period).max()
    k = ((data['收盘'] - low_min) / (high_max - low_min)) * 100
    return k

# 计算薛斯通道（STC）
def schaff_trend_cycle(data, macd_period=3, stochastic_period=3):
    macd = calculate_macd(data)
    k = calculate_stochastic(macd, period=macd_period)
    stc = k.rolling(window=stochastic_period).mean()
    return stc

# 计算STC
df['STC'] = schaff_trend_cycle(df)

# 绘制图形
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['收盘'], label='收盘价', color='blue', alpha=0.5)
plt.plot(df.index, df['STC'], label='薛斯通道 (STC)', color='red')
plt.axhline(y=25, color='green', linestyle='--', label='超卖线 (25)')
plt.axhline(y=75, color='red', linestyle='--', label='超买线 (75)')
plt.xlabel('日期')
plt.ylabel('价格/STC值')
plt.title('双林股份 (300100) 薛斯通道')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# === 中文显示配置 ===
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']  # 设置中文字体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# === 文件读取与预处理 ===
filename = '300100_双林股份_historical_data'
try:
    df = pd.read_csv(f"{filename}.csv", encoding='utf-8')   
# read_csv(f"{filename}.csv", encoding='utf-8') 是 Python 3.6+ 引入的格式化字符串 (f-string) 语法，它可以在字符串中直接插入变量或表达式，并进行计算。等价于filename = "data"
# formatted_string = "{}.csv".format(filename)
except FileNotFoundError:
    raise FileNotFoundError(f"文件 {filename}.csv 不存在，请检查路径！")

# 列名检查与数据预览
print("列名:", df.columns.tolist())
print("\n数据预览:\n", df.head())

# 关键列存在性验证
required_columns = {'日期', '收盘', '最高', '最低'}
if not required_columns.issubset(df.columns):
    missing = required_columns - set(df.columns)
    raise KeyError(f"缺少必要列: {missing}")

# 日期处理
df['日期'] = pd.to_datetime(df['日期'], errors='coerce')
invalid_dates = df[df['日期'].isna()]
if not invalid_dates.empty:
    print(f"\n警告：发现 {len(invalid_dates)} 行无效日期，已删除:")
    print(invalid_dates)
df = df.dropna(subset=['日期']).set_index('日期')

# === 指标计算 ===
def calculate_macd(data, fast=12, slow=26, signal=9):
    """标准MACD计算（快速EMA12, 慢速EMA26, 信号线EMA9）"""
    ema_fast = data['收盘'].ewm(span=fast, adjust=False).mean()
    ema_slow = data['收盘'].ewm(span=slow, adjust=False).mean()
    macd = ema_fast - ema_slow
    signal_line = macd.ewm(span=signal, adjust=False).mean()
    return macd, signal_line

def calculate_stochastic(data, period=14, k_period=3, d_period=3):
    """标准随机指标（%K和%D）"""
    low_min = data['最低'].rolling(period).min()
    high_max = data['最高'].rolling(period).max()
    k = (data['收盘'] - low_min) / (high_max - low_min) * 100
    d = k.rolling(k_period).mean()
    return k, d

def schaff_trend_cycle(data, macd_period=10, stochastic_period=3):
    """薛斯通道（基于MACD与随机指标的复合指标）"""
    macd = calculate_macd(data)[0]  # 获取MACD线
    k, _ = calculate_stochastic(pd.DataFrame({'收盘': macd, '最高': macd, '最低': macd}), 
                               period=macd_period)
    stc = k.rolling(stochastic_period).mean()
    return stc

# 计算指标
df['STC'] = schaff_trend_cycle(df)

# === 可视化优化 ===
plt.figure(figsize=(16, 8))

# 价格曲线
plt.plot(df.index, df['收盘'], label='收盘价', color='#1f77b4', linewidth=1.5, alpha=0.8)

# STC通道
plt.plot(df.index, df['STC'], label='薛斯通道 (STC)', color='#d62728', linewidth=2)

# 超买超卖线
plt.axhline(25, color='#2ca02c', linestyle='--', linewidth=1, label='超卖区 (25)')
plt.axhline(75, color='#ff7f0e', linestyle='--', linewidth=1, label='超买区 (75)')

# 图表装饰
plt.title('双林股份 (300100) 薛斯通道分析', fontsize=16, pad=20)
plt.xlabel('日期', fontsize=12)
plt.ylabel('价格 / STC 值', fontsize=12)
plt.legend(loc='upper left', frameon=True, shadow=True)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator

# ================== 全局配置 ==================
plt.rcParams.update({
    'font.sans-serif': 'Microsoft YaHei',  # 中文字体
    'axes.unicode_minus': False,           # 显示负号
    'figure.dpi': 150                      # 显示分辨率
})

# ================== 数据准备 ==================
def load_and_preprocess(filename):
    """数据加载与预处理"""
    try:
        df = pd.read_csv(f"{filename}.csv", 
                         encoding='utf-8',
                         parse_dates=['日期'],
                         infer_datetime_format=True)
    except FileNotFoundError:
        raise FileNotFoundError(f"文件 {filename}.csv 不存在")
        
    # 日期有效性检查
    date_mask = df['日期'].isna()
    if date_mask.any():
        print(f"发现 {date_mask.sum()} 行无效日期，示例：\n{df[date_mask].head()}")
        df = df[~date_mask]
    
    return df.set_index('日期')

# ================== 技术指标计算 ==================
class TechnicalIndicators:
    @staticmethod
    def weighted_price(df):
        """计算加权价格（修正广播维度）"""
        weights = np.array([2, 1, 1])  # 保持一维
        weighted = (df[['收盘', '最高', '最低']].values * weights).sum(axis=1) / 4
        return pd.Series(weighted, index=df.index)
    
    @staticmethod
    def volatility(df, window=20):
        """计算波动率指标"""
        ma = df['收盘'].rolling(window).mean()
        # 计算收盘价的 window 周期简单移动平均线（SMA），在这里就是20日均线
        return (TechnicalIndicators.weighted_price(df) - ma).abs() / ma
    
    @staticmethod
    def dynamic_ma(df, alpha_series, adjust=False):
        """动态指数移动平均"""
        alpha = alpha_series.values
        weights = (1 - alpha) ** np.arange(len(alpha))[::-1, None]
        weights /= weights.sum(axis=0)
        return (df.values * weights).sum(axis=0)
    
    @staticmethod
    def adx_calculation(df, period=14):
        """平均趋向指数计算"""
        df = df.copy()
        df['TR'] = np.maximum(
            df['最高'] - df['最低'],
            np.maximum(
                abs(df['最高'] - df['收盘'].shift()),
                abs(df['最低'] - df['收盘'].shift())
            )
        )
        
        df['+DM'] = np.where(
            (df['最高'].diff() > df['最低'].shift().diff().abs()),
            df['最高'].diff().clip(lower=0), 0)
        
        df['-DM'] = np.where(
            (df['最低'].shift().diff().abs() > df['最高'].diff()),
            df['最低'].shift().diff().abs().clip(lower=0), 0)
        
        # 平滑处理
        smooth = lambda x: x.ewm(alpha=1/period, adjust=False).mean()
        df['+DI'] = smooth(df['+DM']) / smooth(df['TR']) * 100
        df['-DI'] = smooth(df['-DM']) / smooth(df['TR']) * 100
        df['ADX'] = smooth(abs(df['+DI'] - df['-DI']) / (df['+DI'] + df['-DI']) * 100)
        
        return df[['+DI', '-DI', 'ADX']]

# ================== 主程序 ==================
if __name__ == "__main__":
    # 参数配置
    N = 50
    M = 10
    
    # 加载数据
    filename = '300100_双林股份_historical_data'
    df = load_and_preprocess(filename)
    
    # ========== 指标计算 ==========
    # 加权价格相关指标
    df['AA'] = TechnicalIndicators.weighted_price(df).rolling(5).mean()
    df['静态支撑带'] = df['AA'] * N / 100
    df['静态压力带'] = df['AA'] * (200 - N) / 100
    
    # 波动率相关指标
    df['CC'] = TechnicalIndicators.volatility(df)
    df['DD'] = TechnicalIndicators.dynamic_ma(df['收盘'], df['CC'])
    
    # 通道计算
    df['动态趋势上轨'] = (1 + M/100) * df['DD']
    df['动态趋势下轨'] = (1 - M/100) * df['DD']
    
    # 成交量加权指标
    df['VAR2'] = df['收盘'] * df['成交量']
    df['VAR3'] = TechnicalIndicators.dynamic_ma(df['VAR2'], pd.Series(0.25, index=df.index))
    
    # ADX指标
    adx_df = TechnicalIndicators.adx_calculation(df)
    df = pd.concat([df, adx_df], axis=1)
    
    # ========== 可视化 ==========
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 12), sharex=True, 
                                  gridspec_kw={'height_ratios': [3, 1]})
    
    # 价格与通道
    price_colors = ['#1f77b4', '#2ca02c', '#d62728', '#ff7f0e', '#9467bd']
    labels = ['收盘价', '静态支撑带', '静态压力带', '动态趋势上轨', '动态趋势下轨']
    for i, col in enumerate(['收盘', '静态支撑带', '静态压力带', '动态趋势上轨', '动态趋势下轨']):
        ax1.plot(df.index, df[col], color=price_colors[i], 
                linewidth=1.5 if i==0 else 1.2, 
                alpha=0.8 if i==0 else 0.6, 
                label=labels[i])
    
    # ADX指标
    ax2.fill_between(df.index, df['ADX'], alpha=0.3, color='purple', label='ADX')
    ax2.axhline(25, color='gray', linestyle='--', alpha=0.7)
    ax2.axhline(75, color='gray', linestyle='--', alpha=0.7)
    
    # 图表装饰
    ax1.set_title(f'{filename.split("_")[1]} 技术分析', fontsize=14, pad=20)
    ax1.set_ylabel('价格水平', fontsize=10)
    ax1.legend(loc='upper left', ncol=3)
    ax1.grid(True, alpha=0.3, linestyle=':')
    ax1.xaxis.set_minor_locator(AutoMinorLocator(5))
    
    ax2.set_title('ADX趋势强度指标', fontsize=10, pad=10)
    ax2.set_ylabel('ADX值', fontsize=9)
    ax2.set_ylim(0, 100)
    ax2.grid(True, alpha=0.2)
    
    plt.tight_layout()
    plt.show()
```

```python
# # 原代码
# df['通道1'] = df['AA'] * N / 100
# df['通道2'] = df['AA'] * (200 - N) / 100
# df['通道3'] = (1 + M/100) * df['DD']
# df['通道4'] = (1 - M/100) * df['DD']

# # 修改后
# df['静态支撑带'] = df['AA'] * N / 100        # 原通道1
# df['静态压力带'] = df['AA'] * (200 - N) / 100  # 原通道2
# df['动态趋势上轨'] = (1 + M/100) * df['DD']    # 原通道3
# df['动态趋势下轨'] = (1 - M/100) * df['DD']    # 原通道4


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator

# ================== 全局配置 ==================
plt.rcParams.update({
    'font.sans-serif': 'Microsoft YaHei',  # 中文字体
    'axes.unicode_minus': False,           # 显示负号
    'figure.dpi': 150                      # 显示分辨率
})

# ================== 数据准备 ==================
def load_and_preprocess(filename):
    """数据加载与预处理"""
    try:
        df = pd.read_csv(f"{filename}.csv", 
                         encoding='utf-8',
                         parse_dates=['日期'],
                         infer_datetime_format=True)
    except FileNotFoundError:
        raise FileNotFoundError(f"文件 {filename}.csv 不存在")
        
    # 日期有效性检查
    date_mask = df['日期'].isna()
    if date_mask.any():
        print(f"发现 {date_mask.sum()} 行无效日期，示例：\n{df[date_mask].head()}")
        df = df[~date_mask]
    
    return df.set_index('日期')

# ================== 技术指标计算 ==================
class TechnicalIndicators:
    @staticmethod
    def weighted_price(df):
        """此函数用于计算加权平均价格(AA)，加权平均价=2*收盘价+最高价+最低价"""
        weights = np.array([2, 1, 1])  # 保持一维，权重分配：收盘价权重为 2；最高价权重为 1；最低价权重为 1

        weighted = (df[['收盘', '最高', '最低']].values * weights).sum(axis=1) / 4
        #  df[['收盘', '最高', '最低']]：选取三列数据，得到一个形状为 (n_days, 3) 的矩阵。
        # .values：将 DataFrame 转换为 NumPy 数组，便于高效计算。
        # * weights：通过广播机制，将权重数组应用到每个交易日的数据。
        # [收盘, 最高, 最低] * [2, 1, 1]  = 收盘*2 + 最高*1 + 最低*1
        # .sum(axis=1)：按行（即每个交易日）求和，得到每日的加权总和。
        # / 4：除以权重总和，得到平均值。使结果与原始价格尺度一致。       

        return pd.Series(weighted, index=df.index)
        # 返回一个 Series 对象，索引与原始数据相同。
    
    # 此函数用于计算平均价格（AA）结合股票的收盘价、最高价和最低价，通过赋予不同权重生成综合价格序列。常用于技术分析中平滑价格波动或识别趋势。

    @staticmethod
    def volatility(df, window=20):
        """计算波动率指标CC,这里计算了当前价格与20周期移动平均的相对波动率。"""
        ma = df['收盘'].rolling(window).mean()
        # 计算收盘价的 window 周期简单移动平均线（SMA），在这里ma就是20日均线
        return (TechnicalIndicators.weighted_price(df) - ma).abs() / ma
        # .abs(). 返回一个包含每个元素绝对值的Series/DataFrame
    
    @staticmethod
    def dynamic_ma(df, alpha_series, adjust=False):
        """动态指数移动平均(DMA)"""
        alpha = alpha_series.values
        weights = (1 - alpha) ** np.arange(len(alpha))[::-1, None]
        weights /= weights.sum(axis=0)
        return (df.values * weights).sum(axis=0)
    
    @staticmethod
    def adx_calculation(df, period=14):
        """平均趋向指数计算;平均方向指数(ADX)和相对方向指数(PDI和MDI)"""
        df = df.copy()
        df['TR'] = np.maximum(
            df['最高'] - df['最低'],
            np.maximum(
                abs(df['最高'] - df['收盘'].shift()),
                abs(df['最低'] - df['收盘'].shift())
            )
        )
        
        df['+DM'] = np.where(
            (df['最高'].diff() > df['最低'].shift().diff().abs()),
            df['最高'].diff().clip(lower=0), 0)
        
        df['-DM'] = np.where(
            (df['最低'].shift().diff().abs() > df['最高'].diff()),
            df['最低'].shift().diff().abs().clip(lower=0), 0)
        
        # 平滑处理
        smooth = lambda x: x.ewm(alpha=1/period, adjust=False).mean()
        df['+DI'] = smooth(df['+DM']) / smooth(df['TR']) * 100
        df['-DI'] = smooth(df['-DM']) / smooth(df['TR']) * 100
        df['ADX'] = smooth(abs(df['+DI'] - df['-DI']) / (df['+DI'] + df['-DI']) * 100)
        
        return df[['+DI', '-DI', 'ADX']]

# ================== 主程序 ==================
if __name__ == "__main__":
    # 参数配置
    N = 50
    M = 10
    
    # 加载数据
    filename = '300100_双林股份_historical_data'
    df = load_and_preprocess(filename)
    
    # ========== 指标计算 ==========
    # 加权价格相关指标
    df['AA'] = TechnicalIndicators.weighted_price(df).rolling(5).mean()
    # print(TechnicalIndicators.weighted_price(df).head())
    df['静态支撑带'] = df['AA'] * N / 100                   # 静态支撑带 = AA * N / 100 就是
    df['静态压力带'] = df['AA'] * (200 - N) / 100
    
    # 波动率相关指标
    df['CC'] = TechnicalIndicators.volatility(df)
    df['DD'] = TechnicalIndicators.dynamic_ma(df['收盘'], df['CC'])
    
    # 通道计算
    df['动态趋势上轨'] = (1 + M/100) * df['DD']
    df['动态趋势下轨'] = (1 - M/100) * df['DD']
    
    # 成交量加权指标
    df['VAR2'] = df['收盘'] * df['成交量']
    df['VAR3'] = TechnicalIndicators.dynamic_ma(df['VAR2'], pd.Series(0.25, index=df.index))
    
    # ADX指标
    adx_df = TechnicalIndicators.adx_calculation(df)
    df = pd.concat([df, adx_df], axis=1)
    
    # ========== 可视化 ==========
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 12), sharex=True, 
                                  gridspec_kw={'height_ratios': [3, 1]})
    
    # 价格与通道
    price_colors = ['#1f77b4', '#2ca02c', '#d62728', '#ff7f0e', '#9467bd']
    labels = ['收盘价', '静态支撑带', '静态压力带', '动态趋势上轨', '动态趋势下轨']
    for i, col in enumerate(['收盘', '静态支撑带', '静态压力带', '动态趋势上轨', '动态趋势下轨']):
        ax1.plot(df.index, df[col], color=price_colors[i], 
                linewidth=1.5 if i==0 else 1.2, 
                alpha=0.8 if i==0 else 0.6, 
                label=labels[i])
    
    # ADX指标
    ax2.fill_between(df.index, df['ADX'], alpha=0.3, color='purple', label='ADX')
    ax2.axhline(25, color='gray', linestyle='--', alpha=0.7)
    ax2.axhline(75, color='gray', linestyle='--', alpha=0.7)
    
    # 图表装饰
    ax1.set_title(f'{filename.split("_")[1]} 技术分析', fontsize=14, pad=20)
    ax1.set_ylabel('价格水平', fontsize=10)
    ax1.legend(loc='upper left', ncol=3)
    ax1.grid(True, alpha=0.3, linestyle=':')
    ax1.xaxis.set_minor_locator(AutoMinorLocator(5))
    
    ax2.set_title('ADX趋势强度指标', fontsize=10, pad=10)
    ax2.set_ylabel('ADX值', fontsize=9)
    ax2.set_ylim(0, 100)
    ax2.grid(True, alpha=0.2)
    
    plt.tight_layout()
    plt.show()
```
